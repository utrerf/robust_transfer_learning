```python
import pandas as pd
pd.set_option('display.max_rows', 500)

# pull core results 

# clean dataframe
df = pd.DataFrame()

# read these files
standard_file_list = ['SVHN.csv', 'CIFAR100.csv', 'CIFAR10.csv', 'KMNIST.csv', 'FMNIST.csv', 'MNIST.csv']

# taking only these columns
cols = ['epoch', 'nat_prec1', 'adv_prec1', 'nat_loss', 'adv_loss', 'train_prec1', 'train_loss', 
        'time', 'source_eps', 'target_ds', 'num_training_images', 'unfrozen_blocks','seed']

for file in standard_file_list:
    df = df.append(pd.read_csv(file).loc[:, cols])
    
grouby_cols = ['target_ds', 'unfrozen_blocks', 'num_training_images', 'source_eps', 'seed'] 
df[(df['epoch']==1) & (df['target_ds']=='cifar10')].groupby(by=grouby_cols).count()
```

```python
# pull baselines results 

base_df = pd.read_csv('baselines.csv')
df = df.append(base_df.loc[:, cols])
```

```python
# pull some eps=8 results 
base_df = pd.read_csv('cifar10_svhn_eps8.csv')
df = df.append(base_df.loc[:, cols])
```

```python
by_list = ['target_ds', 'source_eps'] 
table = df.groupby(by=by_list).apply(lambda x: x[x['nat_prec1'] == max(x['nat_prec1'])]).reset_index(drop=True)
table = table.groupby(by=by_list).apply(lambda x: x[x['epoch'] == max(x['epoch'])]).reset_index(drop=True)
display(table.pivot(columns='source_eps', index='target_ds', values='nat_prec1'))
display(table)
```

```python
x = table.pivot(columns='source_eps', index='target_ds', values='nat_prec1')
display(x.style.apply(lambda x: ['font-weight: bold' if v == max(x) else "" for v in x], axis=1))

x = table.pivot(columns='source_eps', index='target_ds', values='epoch')
display(x.style.apply(lambda x: ['font-weight: bold' if v == max(x) else "" for v in x], axis=1))
```

```python
df.to_csv('consolidated_results.csv')
```
